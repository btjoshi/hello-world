# Two interfaces

$ State-machine\  environment $: (known as pyplot) is a complex term for saying that Matplotlib implicitly handles some of the plotting for you. 

$ The\  object-oriented\  approach$: Gives full control over each element of the plot and is the more Pythonic way of interacting with Matplotlib. 
- Most cases can br reproduced with either approach, but the object-oriented approach is explicit and appears easier to determine what is happening.
- much of the code between each approach looks quite similar when making a single plot.

# There is a hierarchy 

The highest and outermost part of a plot is the $ Figure $, contains all the other plotting elements. Typically, you do not interact with it much.

Inside the Figure is the Axes. This is the actual plotting surface that normally referred as ‘plot’. 
- A Figure may contain any number of these Axes. 
- The Axes is a container for all of the other physical pixels that get drawn onto your screen. This includes the x and y axis, lines, text, points, legends, images, etc…

The term Axes is not actually plural and does not mean more than one axis. It literally stands for a single ‘plot’. It’s unfortunate that this fundamental element has a name that is so confusing.

To use embed plots created from matplotlib in a Jupyter Notebook magic command %matplotlib inline is needed. This embeds plots into the output of the cells. Otherwise show method from Figure object needs to be called, which produces a plot in a separate
window.

# Latch on to objects and then call their methods

Everything on our plot is a separate object, may be explicitly referenced by a variable. Once we have a reference to a particular object, we can then modify it by calling methods on it.

There are many objects on Axes that can be referenced, fig, ax, the x and y axis, tick marks, tick labels, and others. 

Calling Axes methods - get_ and set_ methods: Many methods begin with either get_ or set_ followed by the part of the Axes that will get retrieved or modified. 
- title
- xlabel/ylabel
- xlim/ylim
- xticks/yticks
- xticklabels/yticklabels

```python
ax.set_title('My First Matplotlib Graph')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_xlim(0, 5)
ax.set_ylim(-10,50)
ax.set_xticks([1.8, 3.99, 4.4])
ax.set_yticks([-99, -9, -1, 22, 44])
#this forces the change in the limits
#For some reason, the tick labels are hidden in some list-like object. Pass it to the list function
list(ax.get_yticklabels())
```

```python
ax.set_xticklabels(['dog', 'cat', 'snake'])
ax.set_xticks([5,20,40,55,60])
fig
```

#  tick locations and labels & text styles

The tick locations are a completely separate concept than the tick labels. The tick locations are always
numeric and determine where on the plot the tick mark will appear. The tick labels are
the strings that are used on the graph.

The tick labels are defaulted to be a string of the tick location, but can be changed. 

If we eet the tick locations again, these labels are used for the first three ticks, but the last two are empty.

Setting text styles:

Common text properties
- size - Number in “points” where 1 point is defaulted to 1/72nd of an inch
- color - One of the named colors. See the colors API for more.
- backgroundcolor - Same as above
- fontname - Name of font as a string
- rotation - Degree of rotation

```python
ax.set_title('Tests', size=15, color='firebrick', backgroundcolor='lightblue', fontname='Courier New', rotation=-20)
ax.set_xlabel('Stylized Fancy X-lable', size=10, color='indigo', fontname='Times New Roman', rotation=15)
#tick parameters
ax.tick_params(axis='x', color='gold', size=25, width=5, direction='inout', pad=20, top=True)
#tick lable parameters
ax.tick_params(axis='y', labelsize=10, labelcolor='tomato', labelrotation=-20, labelright=True)
fig
```
